```python
# Gerekli kütüphaneler eklendi.

import csv
import datetime

from google.colab import drive
drive.mount('/content/drive/')
```

```python
# Pizza için alt ve üst sınıflar oluşturuldu. Farklı türdeki pizzalara özellikleri verildi.

class Pizza(object):

  def __init__(self,ad,fiyat,aciklama):
    self.ad = ad
    self.fiyat = fiyat
    self.aciklama = aciklama

    def get_description(self):
      return self.aciklama

    def get_cost(self):
      return self.fiyat

klasik = Pizza("Klasik", 65, "Domates sosu, mozzarella, salami")
margherita = Pizza("Margherita",70 ,"Domates sosu, mozarella, fesleğen, zeytinyağı")
turk = Pizza("Türk", 72,"Domates sosu, kaşar peynir, sucuk")
dominos = Pizza("Dominos",68,"Jambon, sucuk, sosis, mozarella, domates sosu, kekik")
```

```python
# Süper sınıf decorator oluşturuldu. Decorator sınıfı, pizza sınıfının özelliklerini kullanarak
# get_description() ve get_cost() yöntemlerini kullanıldı. Sos özellikleri tanımlandı.
class decorator(object):
    
    def __init__(self,ad,fiyat):
      self.ad = ad
      self.fiyat = fiyat

    def get_cost(self):
       return self.component.get_cost() + Pizza.get_cost(self)

    def get_description(self):
       return self.component.get_description() + " " + Pizza.get_description(self)

zeytin = decorator("Zeytin",5)
mantarlar = decorator("Mantar",7)
kecipeyniri = decorator("Keçi Peyniri",12)
et = decorator("Et",15)
sogan = decorator("Soğan",6)
mısır = decorator("Mısır",5)
```

```python
# main fonksiyonu oluşturuldu. Menü kullancıya sunuldu ve giriş istendi.

def main():

  path = '/content/drive/MyDrive/PizzaOrderSystem/Menu.txt'
  dosya = open(path,"r",encoding="utf-8")

  for a in range(0,5):
    print(dosya.readline()) 
  x = input("Pizza seçmek için numarasını yazınız: ")
  for b in range(0,7):
    print(dosya.readline())
  y = input("Sos seçmek için numarasını yazınız: ")

 # Kullanıcının seçimlerini anlamak için pizzalar numaralandırılıp "if elif else" ile kontrol edildi.

  if x == "1":
    x = klasik
  elif x == "2":
     x = margherita
  elif x == "3":
     x = turk
  elif x == "4":
     x = dominos
  else:
     print("Geçerli değer giriniz!")

  if y == "11":
   y = zeytin
  elif y == "12":
   y = mantarlar
  elif y == "13":
    y = kecipeyniri
  elif y == "14":
   y = et
  elif y == "15":
    y = sogan
  elif y == "16":
   y = mısır
  else:
     print("Geçerli değer giriniz!")

  # Kullanıcıdan bilgiler istendi ve bu bilgiler kontrol edildi.

  print("Ödeyeceğiniz toplam fiyat: ")
  print(x.fiyat + y.fiyat)
 
  isim = input("İsminizi giriniz: ")
  
  tc = input("11 haneli TC No giriniz: ")
  while len(tc) != 11:
    tc = input("11 haneli TC No giriniz: ")

  ccNo = input("Kredi kartı no giriniz: ")
  while len(ccNo) != 16:
    print("Kredi kartı numaranız 16 haneli olmalıdır.")
    ccNo = input("Kredi kartı no giriniz: ")

  ccPass = input("Kredi kartı şifrenizi giriniz: ")
  while len(ccPass) != 4:
    ccPass = input("Kredi kartı şifrenizi giriniz: ")

  comment = input("Açıklama giriniz: ")
  
  an = datetime.datetime.now()
  tarih = datetime.datetime.strftime(an, '%c')
 
  print("     ------FIS-------")
  print("Pizza içeriği: ",x.aciklama,",",y.ad )
  print("Seçilen ürünler --- fiyatlar")
  print(x.ad,"             ",x.fiyat)
  print(y.ad,"             ",y.fiyat)
  print("Toplam fiyat         ",x.fiyat+y.fiyat)
  print("\n")
  print(isim,"    TC: ",tc)
  print("Kart No:   ",ccNo)

# Sipariş bilgileri csv dosyasına kaydedildi.

  row = {'user_name': isim, 'user_id 	':tc, 'credit_id': ccNo, 'comment': comment, 'date': tarih, 'password': ccPass}

  with open('/content/drive/MyDrive/PizzaOrderSystem/Orders_Database.csv', 'a', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=row.keys())  
    writer.writerow(row)

if __name__== "__main__":
  main()
```
